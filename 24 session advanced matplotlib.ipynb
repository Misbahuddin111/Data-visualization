```python
batters = pd.read_csv('Datasets/batter.csv')
sample_df = batters.sample(25)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
z = np.sin(xx) + np.log(xx)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour plots
fig = plt.Figure()
ax = plt.subplot()
p=ax.contour(xx,yy,z,cmap="viridis")
fig.colorbar(p)
fig.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(-5, 5, 100)  # Create 1D arrays for x and y
y = np.linspace(-5, 5, 100)
xx, yy = np.meshgrid(x, y)   # Create 2D meshgrid for contour plot
z = np.sin(np.sqrt(xx**2 + yy**2))  # Example function for z

# Create the plot
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot()

# Create contour plot
p = ax.contour(xx, yy, z, cmap='viridis')

# Add a colorbar
fig.colorbar(p)

# Display the plot
plt.show()
```
