Q-1: Using Gapminder Data
Create a scatter plot of 'gdpPercap' against 'lifeExp' for the year 2007, with the size of the markers determined by 'pop' and the color determined by 'continent'.

```python
import plotly.express as px
df = px.data.gapminder()
df.head(2)
temp_df=df[df['year'].isin([2007])]
```

```python
sns.scatterplot(data=temp_df,x='gdpPercap',y='lifeExp',hue='continent',size='pop')
```

Q2 Using the "flights" dataset that comes with seaborn, create a heatmap that shows the average number of passengers per month for each year.

```python
df = sns.load_dataset('flights')
df.head(2)
```

```python
temp = df.pivot_table(index='year',columns='month',values='passengers',aggfunc='mean')
sns.heatmap(data=temp)
```

Using the seaborn's flight dataset, create a clustermap to visualize the relationship between the number of passengers, months, and year.


```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('flights')
temp = df.pivot_table(index='year', columns='month', values='passengers', aggfunc='mean')
sns.clustermap(data=temp, annot=True, fmt='.0f', cmap='YlGnBu', figsize=(10, 8), 
               cbar_kws={'label': 'Average Passengers'})
plt.title('Clustermap of Passengers by Year and Month', fontsize=14, pad=20)
plt.show()
```
