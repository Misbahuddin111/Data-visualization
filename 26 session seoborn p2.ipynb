difference betweeen box plot and violin plot
Key Differences
Box Plot: Simple, shows summary stats (median, quartiles, outliers).
Violin Plot: More detailed, shows full distribution shape via KDE.
When to Use:
Box plot for quick comparisons or outlier detection.
Violin plot for understanding distribution density or shape.

pair grid
Use case: When you need fine-grained control over the types
 of plots or customizations in a pairwise plot grid.

```python
# pair grid
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 4, 3, 2, 1]
})
g = sns.PairGrid(df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.barplot)
```

pair plot
Definition: pairplot is a higher-level, convenience function in Seaborn that creates a grid of plots (similar to PairGrid) to visualize pairwise relationships in a dataset with less manual configuration.
Use case: When you want a quick, out-of-the-box visualization of pairwise relationships without extensive customization.

```python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 4, 3, 2, 1]
})
sns.pairplot(df)
```

When to Use
Use pairplot for quick, default visualizations during exploratory data analysis.
Use PairGrid when you need specific plot types (e.g., regression plots, custom functions) or advanced customizations.
Both are powerful for understanding relationships between multiple variables in a dataset!


```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=tips,)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

Ease of Use: Simple, single-function call. Ideal for quick exploratory data analysis.
Components:
A central plot (e.g., scatter, hexbin, KDE, regression).
Marginal plots (histograms or KDEs) for each variable.
Customization: Limited to parameters like kind (scatter, kde, hex, reg, resid), hue, height, etc. For deeper customization, you need to access the underlying JointGrid object.

```python
# joinplot
import seaborn as sns
import pandas as pd

# Sample data
df = sns.load_dataset("penguins")
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", kind="scatter", hue="species")
```

Components:
A grid with a central axes (ax_joint) and marginal axes (ax_marg_x, ax_marg_y).
You explicitly call plotting functions (e.g., sns.scatterplot, sns.histplot) on these axes.
Customization: Highly flexible. You can use any plotting function compatible with Matplotlib axes, combine multiple plot types, and tweak aesthetics extensively.

```python
import seaborn as sns
import pandas as pd

# Sample data
df = sns.load_dataset("penguins")
g = sns.JointGrid(data=df, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.scatterplot, hue=df["species"])  # Central scatter plot
g.plot_marginals(sns.histplot, kde=True)  # Marginal histograms with KDE
```

When to Use
Use jointplot for quick, standardized bivariate plots with minimal setup (e.g., exploratory analysis).
Use JointGrid when you need custom plot types, layered visualizations, or specific tweaks (e.g., publication-quality figures).
