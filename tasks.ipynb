```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

Task: Create a bar plot showing the average tip amount for each day of the week.

```python
sns.catplot(data=tips,x='day',y='tip',kind='bar',ci=None)
plt.title('average tip')
plt.xlabel('tip given')
plt.ylabel('each day')
plt.show()
```

ask: Create a scatter plot to visualize the relationship between sepal length and sepal width, with points colored by species.

```python
iris.head(2)
sns.relplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
```

Task: Create a box plot to show the distribution of body mass (body_mass_g) across different penguin species.

```python
penguins = sns.load_dataset('penguins')
penguins.head(3)
sns.catplot(data=penguins,x='species',y='body_mass_g',kind='box')
```

Task: Create a pair plot to visualize relationships between all numerical variables (sepal length, sepal width, petal length, petal width), with points colored by species and a kernel density estimate on the diagonal.

```python
sns.pairplot(data=iris,hue='species',diag_kind='kde',palette='Set2')
```

Task: Create a heatmap to show the number of passengers (passengers) by month and year.

```python
flights = sns.load_dataset('flights')
temp_df = pd.pivot_table(data=flights,index='year',columns='month',values='passengers')
```

```python
sns.heatmap(data=temp_df,cmap='YlGnBu')
```
